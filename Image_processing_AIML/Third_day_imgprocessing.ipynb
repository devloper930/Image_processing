```python
import cv2
vid = cv2.VideoCapture(0)
while True:
    flag, img = vid.read()
    # gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img2 = cv2.threshold(img, 75, 250, cv2.THRESH_BINARY)
    if flag:
        # cv2.imshow("web_image", img)
        # cv2.imshow("web_image",gray)
        cv2.imshow("web_image", img2[1])
        key = cv2.waitKey(1)
        if key == ord("q"):
            break
    else:
        break
cv2.destroyAllWindows()
vid.release()
```

face detection

```python
# to draw a rectangle

import cv2
fd=cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
# video using web cam
vid = cv2.VideoCapture(0)

# main loop to read and show image until we break the loop
while True:
    flag, img = vid.read()
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # if flag is true then show image
    if flag:
        faces = fd.detectMultiScale(gray,1.1,5)

        for x,y,w,h in faces:
            # draw a rectangle om image from point 1 to point 2
            cv2.rectangle(gray, pt1=(x, y), pt2=(x+w,y+h), color=(255, 0, 0), thickness=3)
        
        # cv2.imshow("web_image", img)
        cv2.imshow("web_image",gray)
        key = cv2.waitKey(1)
        if key == ord("q"):
            break
    else:
        break
cv2.destroyAllWindows()
vid.release()
```
