```python
# ord("q")
def image_show(image_path):
    img = cv2.imread(image_path)
    cv2.imshow(image_path.split(".")[0], img)
    
    while True:
        key = cv2.waitKey(0)
        if key == ord("q"):
            cv2.destroyAllWindows()
            break
image_show("spider-mon.jpeg")

```
